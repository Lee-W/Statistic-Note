# Discrete Probability Distribution

- $ E(x) = \mu = \sum {x \cdot f(x)} $
    - x: 隨機變數
    - f(x): 機率
   
   
- $ Var(x) = \sigma^2 
\\= \sum {(x-u)^2 f(x)}
\\ = E(x^2) - [E(x)]^2$ 

```python
# variance
np.var([1, 2, 3])
```

## Binominal Distribution

1. n個試驗
2. 只有成功失敗
3. 機率不會隨著試驗改變
4. 試驗相互獨立

如果只符合2, 3, 4 $ \to $ 柏努利事件(Bernoulli Process)

- $ f(x) = \frac {n!} {x! (n-x)!} p^x (1-p)^{(n-x)} $
- $ E(x) = n p $
- $ Var(x) = n p (1 - p) $

```python
n, p = 1000, .6
x = np.random.binomial(n=1000, p=.6, size=1000)
plt.plot(x);
```

## Poision Distribution
1. 只要在區間長度相同，機率即相同
2. 任一區間發生與否不影響其他區間是否發生

- $ f(x) = \frac {\mu \cdot {e ^ {-\mu}}} {x!} $
- $ \mu = \sigma ^ 2 $   (for Poisson Distribution)

```python
interval = 100
x = np.random.poisson(interval, size=1000)
plt.hist(x);
```

## Hypergeometric Distribution
1. 取後不放回 $ \Rightarrow $ 各事件相互影響 $ \Rightarrow $ 不獨立

2. $ f(x) = \frac {{r \choose x} \cdot {(N-r) \choose (n-x)}} {N \choose n}$   (for $ 0 \leqslant x \leqslant r $)
3. $ E(x) = n (\frac{r}{N}) \hspace{0.7cm} for \hspace{0.2cm} 0 \leq x \leq r$
4. $ var(x) = n (\frac{r}{N}) (1-\frac{r}{N})(\frac{N-n}{N-1})$

```python
x = np.random.hypergeometric(ngood=100, nbad=50, nsample=100, size=100)
plt.plot(x);
```
